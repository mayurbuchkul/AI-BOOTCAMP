```python
#1. Plot the number of visitors over different days (Line Plot).#
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5']
visitors = np.random.randint(50000, 200000, size=5)
plt.plot(days, visitors)
plt.xlabel('Days')
plt.ylabel('visitors')
plt.title('Line Plot')
plt.grid(True)
plt.show()
```

```python
#2. Show the distribution of participants by city (Bar Chart).
cities = ['Prayagraj', 'Varanasi', 'Lucknow', 'Kanpur', 'Agra']
participants = np.random.randint(5000, 65000, size=5)
df_visitors = pd.DataFrame({'Day': days, 'Visitors': visitors})
df_participants = pd.DataFrame({'City': cities, 'Participants': participants})
plt.figure(figsize=(8, 5))
plt.bar(df_participants['City'], df_participants['Participants'])
plt.xlabel('Cities')
plt.ylabel('Number of Participants')
plt.title('Participants from Different Cities - Khubh Mela')
plt.show()
```

```python
#3. Use legends, gridlines, and save the plot.
plt.figure(figsize=(8, 5))
days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5']
plt.plot(days, visitors, label='Visitors Trend', linestyle='-', marker='s')
plt.plot(days, participants, label='Participants Trend', linestyle='--', marker='o')

plt.xlabel('Days')
plt.ylabel('Visitors and Participents Count')
plt.title('Visitors & Participants Trend - Khubh Mela')
plt.legend()
plt.savefig('khubh_mela.png')
plt.grid(True)
plt.show()

```
